```python
# Imports
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Basic Plots
# 1. Basic line graph
plt.figure()
plt.plot([1, 2, 3, 4])
plt.title("Basic Line Graph")
plt.show()

# 2. Two lines on same graph
plt.figure()
plt.plot([1, 2, 3], [4, 5, 6], label="Line 1")
plt.plot([1, 2, 3], [6, 5, 4], label="Line 2")
plt.legend()
plt.title("Two Lines")
plt.show()

# 3. Change line color
plt.figure()
plt.plot([1, 2, 3], color='red')
plt.title("Red Line")
plt.show()

# 4. Use markers
plt.figure()
plt.plot([1, 2, 3], marker='o')
plt.title("Line with Markers")
plt.show()

# 5. Change line style
plt.figure()
plt.plot([1, 2, 3], linestyle='--')
plt.title("Dashed Line")
plt.show()

# Labels & Legends
# 6. Add title
plt.figure()
plt.plot([1, 2, 3])
plt.title("Plot Title")
plt.show()

# 7. Label axes
plt.figure()
plt.plot([1, 2, 3])
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("Labeled Axes")
plt.show()

# 8. Add legend
plt.figure()
plt.plot([1, 2, 3], label="Data")
plt.legend()
plt.title("Legend Example")
plt.show()

# Bar Charts
# 9. Simple bar chart
plt.figure()
plt.bar(['A', 'B', 'C'], [10, 20, 15])
plt.title("Bar Chart")
plt.show()

# 10. Change bar colors
plt.figure()
plt.bar(['A', 'B', 'C'], [10, 20, 15], color=['red', 'green', 'blue'])
plt.title("Colored Bars")
plt.show()

# 11. Horizontal bar chart
plt.figure()
plt.barh(['A', 'B', 'C'], [10, 20, 15])
plt.title("Horizontal Bar Chart")
plt.show()

# Histograms
# 12. Histogram
data = np.random.randn(1000)
plt.figure()
plt.hist(data)
plt.title("Histogram")
plt.show()

# 13. Change number of bins
plt.figure()
plt.hist(data, bins=30)
plt.title("Histogram with 30 Bins")
plt.show()

# Customization
# 14. Add gridlines
plt.figure()
plt.plot([1, 2, 3])
plt.grid(True)
plt.title("Gridlines")
plt.show()

# 15. Save plot as image
plt.figure()
plt.plot([1, 2, 3])
plt.title("Saved Plot")
plt.savefig("saved_plot.png")
plt.close()

# 16. Customize tick marks
plt.figure()
plt.plot([1, 2, 3])
plt.xticks([0, 1, 2, 3])
plt.yticks([0, 1, 2, 3])
plt.title("Custom Ticks")
plt.show()

# 17. Rotate x-axis labels
plt.figure()
plt.bar(['Jan', 'Feb', 'Mar'], [10, 20, 15])
plt.xticks(rotation=45)
plt.title("Rotated Labels")
plt.show()

# 18. Add annotations
plt.figure()
plt.plot([1, 2, 3])
plt.annotate("Peak", xy=(2, 2), xytext=(2, 2.5), arrowprops=dict(arrowstyle="->"))
plt.title("Annotation")
plt.show()

# 19. Fill area under line
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.figure()
plt.plot(x, y)
plt.fill_between(x, y, alpha=0.3)
plt.title("Filled Area")
plt.show()

# 20. Different figure size
plt.figure(figsize=(8, 4))
plt.plot([1, 2, 3])
plt.title("Custom Figure Size")
plt.show()

# 21. Customize legend location
plt.figure()
plt.plot([1, 2, 3], label="Data")
plt.legend(loc='lower right')
plt.title("Legend Location")
plt.show()

# 22. Font styles and sizes
plt.figure()
plt.plot([1, 2, 3])
plt.title("Styled Title", fontsize=14, fontweight='bold')
plt.xlabel("X", fontsize=12)
plt.ylabel("Y", fontsize=12)
plt.show()

# 23. Change background color
fig, ax = plt.subplots()
ax.set_facecolor('lightgray')
ax.plot([1, 2, 3])
plt.title("Background Color")
plt.show()

# 24. Use plt.style
plt.style.use('ggplot')
plt.figure()
plt.plot([1, 2, 3])
plt.title("Styled Plot")
plt.show()

# 25. Set axis limits
plt.figure()
plt.plot([1, 2, 3])
plt.xlim(0, 5)
plt.ylim(0, 5)
plt.title("Axis Limits")
plt.show()

# 26. Hide axis ticks and labels
plt.figure()
plt.plot([1, 2, 3])
plt.xticks([])
plt.yticks([])
plt.title("No Ticks or Labels")
plt.show()

# 27. Customize plot borders/spines
fig, ax = plt.subplots()
ax.plot([1, 2, 3])
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.title("Custom Spines")
plt.show()

# Pie Charts
# 28. Pie chart
plt.figure()
plt.pie([30, 40, 30])
plt.title("Pie Chart")
plt.show()

# 29. Pie chart with labels
plt.figure()
plt.pie([30, 40, 30], labels=['A', 'B', 'C'])
plt.title("Pie Chart with Labels")
plt.show()

# 30. Pie chart with percentages
plt.figure()
plt.pie([30, 40, 30], labels=['A', 'B', 'C'], autopct='%1.1f%%')
plt.title("Pie Chart with Percentages")
plt.show()

# 31. Donut chart
plt.figure()
wedges, texts = plt.pie([30, 40, 30])
plt.gca().add_artist(plt.Circle((0, 0), 0.5, color='white'))
plt.title("Donut Chart")
plt.show()

# Scatter Plots
# 32. Scatter plot
plt.figure()
plt.scatter([1, 2, 3], [4, 5, 6])
plt.title("Scatter Plot")
plt.show()

# 33. Change marker size
plt.figure()
plt.scatter([1, 2, 3], [4, 5, 6], s=100)
plt.title("Large Markers")
plt.show()

# 34. Change marker color
plt.figure()
plt.scatter([1, 2, 3], [4, 5, 6], color='green')
plt.title("Colored Markers")
plt.show()

# 35. Multiple scatter plots
plt.figure()
plt.scatter([1, 2, 3], [4, 5, 6], label="Group 1")
plt.scatter([1, 2, 3], [6, 5, 4], label="Group 2")
plt.legend()
plt.title("Multiple Scatter Plots")
plt.show()

# Subplots
# 36. Two plots side-by-side
plt.figure()
plt.subplot(1, 2, 1)
plt.plot([1, 2, 3])
plt.title("Left Plot")
plt.subplot(1, 2, 2)
plt.plot([3, 2, 1])
plt.title("Right Plot")
plt.tight_layout()
plt.show()

# 37. Four plots in 2x2 grid
plt.figure()
for i in range(1, 5):
    plt.subplot(2, 2, i)
    plt.plot([i, i+1, i+2])
    plt.title(f"Plot {i}")
plt.tight_layout()
plt.show()

# Advanced Plots
# 38. Sine wave
x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)
plt.figure()
plt.plot(x, y)
plt.title("Sine Wave")
plt.show()

# 39. Dual y-axes
x = np.arange(0, 10)
y1 = x**2
y2 = np.sqrt(x)
fig, ax1 = plt.subplots()
ax2 = ax1.twinx()
ax1.plot(x, y1, 'g-')
ax2.plot(x, y2, 'b-')
plt.title("Dual Y-Axes")
plt.show()

# 40. 3D plot
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_title("3D Surface Plot")
plt.show()
```
